**Train Test Split**

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
# split
X, y = df.drop(['benign_0__mal_1'],axis=1), df.benign_0__mal_1
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
# scale
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train, X_test = scaler.transform(X_train), scaler.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation,Dropout
# fit model
model = Sequential()
model.add(Dense(units=30,activation='relu'))
model.add(Dense(units=15,activation='relu'))
model.add(Dense(units=1,activation='sigmoid')) # because binary classification, output layer is specified
model.compile(loss='binary_crossentropy', optimizer='adam')
model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test, y_test), verbose=1)
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

Avoid overfitting: **use uearly stopping**

```python
from tensorflow.keras.callbacks import EarlyStopping
model = Sequential()
model.add(Dense(units=30,activation='relu'))
model.add(Dense(units=15,activation='relu'))
model.add(Dense(units=1,activation='sigmoid')) # because binary classification, output layer is specified
model.compile(loss='binary_crossentropy', optimizer='adam')
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=25)
# help(EarlyStopping)
model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test, y_test), verbose=1, callbacks=[early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

Avoid overfitting: **Adding in DropOut layers**

```python
model = Sequential()
model.add(Dense(units=30,activation='relu'))
model.add(Dropout(0.5)) # 50% nodes turned off
model.add(Dense(units=15,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(units=1,activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam')
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=25)
# help(EarlyStopping)
model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test, y_test), verbose=1, callbacks=[early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

```python
# Evaluation
from sklearn.metrics import classification_report,confusion_matrix
pred = model.predict(X_test)
pred = np.where(pred > 0.5, 1,0)
```

```python
print(classification_report(y_test,pred))
print(confusion_matrix(y_test,pred))
```
